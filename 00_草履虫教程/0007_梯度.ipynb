```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 连续性


x = np.arange(-20,20)
y = 0.7*x**2
z=-0.5*x**2

plt.grid() # 网格


# X轴 y轴 加粗
plt.axhline(y=0, lw=0.7, color='k')
plt.axvline(x=0, lw=0.7, color='k')

plt.plot(x,y,z)
plt.show()
```
