```python
import numpy as np
import matplotlib.pyplot as plt
```

0.1 简单的函数


```python
# 极限的函数 举例


x = np.arange(-100,100)
y = 1-x*np.sin(1/x)


plt.grid() # 网格
plt.plot(x,y)
plt.show()
```

```python
from tkinter import _XYScrollCommand


def test(x):
    return 1/x


x = np.arange(-100,100)
y = test(x)

# 网格
plt.grid() 

# X轴 y轴 加粗
plt.axhline(y=0, lw=0.7, color='k')
plt.axvline(x=0, lw=0.7, color='k')

plt.plot(x,y)
plt.show()
```
