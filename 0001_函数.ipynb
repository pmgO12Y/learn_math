```python
import numpy as np
import matplotlib.pyplot as plt
```

0.1 简单的函数


```python

def my_func(x):
    # 新建一个函数 3x**2+2x+1
    return 3 * x ** 2 + 2 * x + 1


# x = np.linspace(-1000000, 1000000)
x=np.arange(-2, 2,0.01)
y = my_func(x)

# 生成X轴和Y轴函数图像
plt.plot(x, y)
```

```python
# # 三角函数
import math
def my_sin(x):
    return np.sin(x)
x = np.linspace(-2, 2)
y= my_sin(x)
plt.plot(x, y)

```

0.2 分段函数

```python
# 分段函数
def my_分段(x):
    if 0 <= x <= 1:
        return 3 * x
    elif 1 < x <= 3:
        return 3
    elif 3 < x <= 5:
        return x



x = np.linspace(0,5)
y = np.vectorize(my_分段)
y=y(x)
plt.plot(x, y)
```

```python
# 分段函数 numpy方法 np.select
# np.select([x < 0, x == 0, x > 0], [x * 3, 3, x])

np.select(([0 <= x <= 1, 1 < x <= 3, 3 < x <= 5], [3 * x, 3, x]))
```
